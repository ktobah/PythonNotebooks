# Reading the csv file & cleaning the data

```python
recent_grads = pd.read_csv("recent-grads.csv")
print(recent_grads.iloc[0])
```

# Print the head and tail of the dataset

```python
print(recent_grads.head())
print(recent_grads.tail())
```

# Generate a statistical descritpion of the dataset

```python
recent_grads.describe()
```

# Removing rows with missing data

```python
raw_data_count = len(recent_grads)
raw_data_count
```

```python
recent_grads = recent_grads.dropna()
cleared_data_count = len(recent_grads)
cleared_data_count
```

Note: only one row containing missing values was deleted.

# Scatter plots to compare different columns

## "Sample_size" and "Median"

```python
recent_grads.plot(x="Sample_size", y="Median", kind="scatter", title="Sample_size vs. Median")
```

## "Sample_size" and "Unemployment_rate"

```python
recent_grads.plot(x="Sample_size", y="Unemployment_rate", kind="scatter", title="Sample_size vs. Unemployment")
```

## "Full_time" and "Median"

```python
recent_grads.plot(x="Full_time", y="Median", kind="scatter", title="Full time vs. Median")
```

## Observation:
There is no apparent link between the number of full-time employees and the median salary. We notice sometimes for almost the same number of employees (full-time), we get higher and lower median.  

## "ShareWomen" and "Unemployment_rate"

```python
recent_grads.plot(x="ShareWomen", y="Unemployment_rate", kind="scatter", title="ShareWomen vs. Unemployment")
```

## "Men" and "Median"

```python
recent_grads.plot(x="Men", y="Median", kind="scatter", title="Men vs. Median")
```

## "Women" and "Median"

```python
recent_grads.plot(x="Women", y="Median", kind="scatter", title="Women vs. Median")
```

## Observation:
From the plot above we conclude that students from majors where the majority are female don't make more money. The data is condensed around a specific area (less than 80000), and there are abvious outliers points.

```python
recent_grads.plot(x="Rank", y="Median", kind="scatter", title="Rank vs. Median")
```

## Observation:
We notice that when the rank of the major increases, the median salary decreases, which proves that students from popular majors make more money.

This scatter plot is a strong negative non-linear plot.
