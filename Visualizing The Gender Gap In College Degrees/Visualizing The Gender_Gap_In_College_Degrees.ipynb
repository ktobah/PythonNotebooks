# Improving data visualization charts
In this project, we'll focus on visualizing the gender gap across a set of college degrees.
## STEM degrees:

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

fig = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    # Draw the two lines corresponding to each column
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    # Disable the spines to improve data-ink
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    # Make the x and y ranges identical for all plots
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    # Set a title for each individual plot
    ax.set_title(stem_cats[sp])
    # Turn off the ticks to improve data-ink
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    # Add text labels to the lines for the first and last plots only
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
plt.show()
```

## All degrees
We have seventeen degrees in total, so we'll use a 6 rows by 3 columns subplot grid. We can then group the degrees into STEM, liberal arts, and other, as follows:

```python
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
```

## Plotting
To make plotting easier, we use the formula 'ax = fig.add_subplot(6,3,3*sp+step)'

```python
fig = plt.figure(figsize=(16, 20))
for sp in range(0,6):
    ax = fig.add_subplot(6,3,3*sp+1)
    # Draw the two lines corresponding to each column
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    # Disable the spines to improve data-ink
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    # Make the x and y ranges identical for all plots
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    # Set a title for each individual plot
    ax.set_title(stem_cats[sp])
    # Turn off the ticks to improve data-ink
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    # Add text labels to the lines for the first and last plots only
    if sp == 0:
        ax.text(2003, 85, 'Men')
        ax.text(2005, 10, 'Women')
    elif sp == 5:
        ax.text(2005, 87, 'Men')
        ax.text(2003, 7, 'Women')

for sp in range(0,5):
    ax = fig.add_subplot(6,3,3*sp+2)
    # Draw the two lines corresponding to each column
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    # Disable the spines to improve data-ink
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    # Make the x and y ranges identical for all plots
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    # Set a title for each individual plot
    ax.set_title(lib_arts_cats[sp])
    # Turn off the ticks to improve data-ink
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    # Add text labels to the lines for the first and last plots only
    if sp == 0:
        ax.text(2003, 78, 'Men')
        ax.text(2005, 18, 'Women')

for sp in range(0,6):
    ax = fig.add_subplot(6,3,3*sp+3)
    # Draw the two lines corresponding to each column
    ax.plot(women_degrees['Year'], women_degrees[other_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    # Disable the spines to improve data-ink
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    # Make the x and y ranges identical for all plots
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    # Set a title for each individual plot
    ax.set_title(other_cats[sp])
    # Turn off the ticks to improve data-ink
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    # Add text labels to the lines for the first and last plots only
    if sp == 0:
        ax.text(2003, 90, 'Men')
        ax.text(2005, 5, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2003, 30, 'Women')

plt.show()
```

## Observation:
We notice that the titles of each plot are overlaping with the x axis of the plot right above it. We can fix this by removing the x axis labels from each plot except the the ones in the bottom. Also, we can make the plots more clearer by tweaking the y axis to only show the values 0 and 100. The code becomes:

```python
fig = plt.figure(figsize=(16, 20))
for sp in range(0,6):
    ax = fig.add_subplot(6,3,3*sp+1)
    # Draw the two lines corresponding to each column
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    # Disable the spines to improve data-ink
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    # Make the x and y ranges identical for all plots
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    # Set y axis to show only 0 and 100
    ax.set_yticks([0,100])
    # Add a horizontal line at y=50 to make reading the data easier
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    # Set a title for each individual plot
    ax.set_title(stem_cats[sp])
    # Turn off the ticks to improve data-ink
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')
    # Add text labels to the lines for the first and last plots only
    if sp == 0:
        ax.text(2003, 85, 'Men')
        ax.text(2005, 10, 'Women')
    elif sp == 5:
        ax.text(2005, 87, 'Men')
        ax.text(2003, 7, 'Women')
# Show the x axis labels only on the bottom plot
ax.tick_params(labelbottom='on')

for sp in range(0,5):
    ax = fig.add_subplot(6,3,3*sp+2)
    # Draw the two lines corresponding to each column
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    # Disable the spines to improve data-ink
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    # Make the x and y ranges identical for all plots
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    # Set y axis to show only 0 and 100
    ax.set_yticks([0,100])
    # Add a horizontal line at y=50 to make reading the data easier
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    # Set a title for each individual plot
    ax.set_title(lib_arts_cats[sp])
    # Turn off the ticks to improve data-ink
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')
    # Add text labels to the lines for the first and last plots only
    if sp == 0:
        ax.text(2003, 78, 'Men')
        ax.text(2005, 18, 'Women')
ax.tick_params(labelbottom='on')

for sp in range(0,6):
    ax = fig.add_subplot(6,3,3*sp+3)
    # Draw the two lines corresponding to each column
    ax.plot(women_degrees['Year'], women_degrees[other_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    # Disable the spines to improve data-ink
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    # Make the x and y ranges identical for all plots
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    # Set y axis to show only 0 and 100
    ax.set_yticks([0,100])
    # Add a horizontal line at y=50 to make reading the data easier
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    # Set a title for each individual plot
    ax.set_title(other_cats[sp])
    # Turn off the ticks to improve data-ink
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')
    # Add text labels to the lines for the first and last plots only
    if sp == 0:
        ax.text(2003, 90, 'Men')
        ax.text(2005, 5, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2003, 30, 'Women')
ax.tick_params(labelbottom='on')

plt.show()
```
